```python
import pandas as pd
import numpy as np
import math
data = pd.read_csv('./citrus.csv')
data
```

```python
class KNN():

    def split(self,data, p = 0.75):
        
        tr_data = data.sample(frac = p, random_state = 100)
        ts_data = data.drop(tr_data.index)

        x_tr = tr_data.iloc[:,[0]]
        y_tr = tr_data.iloc[:,[1,2,3,4,5]]
        x_ts = ts_data.iloc[:,[0]]
        y_ts = ts_data.iloc[:,[1,2,3,4,5]]
        
        return x_tr, y_tr, x_ts, y_ts
        
    def euclidean_distance(self, data ,k):

        class_result = []
        x_tr_l = data[0].values
        x_ts_l = data[2].valuse
        for i in range(0,len(x_ts_l)):
            df = x_tr_l - x_ts_l[i]
            df = df ** 2
            class_result.append(np.sqrt(df)) 
    
        return class_result[:k]
        
```

```python
a = KNN()
b = a.split(data)
```
