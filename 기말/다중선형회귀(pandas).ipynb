```python
import pandas as pd
import numpy as np

data = pd.read_csv('./data.txt',delimiter = '\t')
```

```python
class multipul_leaner_regression():
    
    def __init__(self, w = 0, inter = 1):
        self.w = w
        self.inter = inter
    
    def preprocessing(self, data):
        if self.inter == 1:
            data['y절편'] = 1
        data.drop(['Id'],axis = 1, inplace = True)
        return data

    def slicing(self, x_array, p):

        tr_data = x_array.sample(frac = p, random_state = 100)
        
        ts_data = x_array.drop(tr_data.index)
        
        return tr_data, ts_data
    
    def learn(self, data):
        x = []
        y = []
        array = data.values
        for i in range(len(array)):
            y.append(array[i][0])
            x.append(array[i][1:])
        x = np.array(x)
        y = np.array(y)
        self.w = np.linalg.inv(x.T @ x) @ x.T @ y
        return self.w
            
    def test(self, data):
        array = data.values
        x = []
        y = []
        pred_y = []
        pred_y = np.array(pred_y)
        
        for i in range(len(array)):
            x.append(array[i][1:])
            y.append(array[i][0])
        
        x = np.array(x)
        y = np.array(y)

        pred_y = np.dot(x, self.w)
        pred_y = pd.DataFrame(pred_y)
        y = pd.DataFrame(y)
        
        test_data = pd.concat([pred_y, y], axis = 1)
        test_data.columns = ['예측값','실제값']
        return test_data
            
class Metric:
    def MSE(self, data):
        array = data.values
        pred_y = []
        y = []
        for i in range(len(array)):
            pred_y.append(array[i][0])
            y.append(array[i][1])
        pred_y = np.array(pred_y)
        y = np.array(y)
        mse_e = sum((y - pred_y)**2)/len(y)
        return mse_e
    
    def MAPE(self, data):
        array = data.values
        pred_y = []
        y = []
        
        for i in range(len(array)):
            pred_y.append(array[i][0])
            y.append(array[i][1])
        
        pred_y = np.array(pred_y)
        y = np.array(y)
        mape_e = sum(abs(y - pred_y)/abs(y))/len(y) * 100
        return mape_e
    
```

```python
l = multipul_leaner_regression()
m = Metric()

p = 0.75 #(75 : 25)

data_pre = l.preprocessing(data) #전처리
slicing = l.slicing(data_pre, p) #training test파일 분할
```

```python
w = l.learn(slicing[0]) # w 값 학습
t = l.test(slicing[1]) # 실제값과 예측값 비교
t #예측값과 실제값 비교 dataframe
```

```python
mse = m.MSE(t) #MSE
mape = m.MAPE(t) #MAPE
print('MSE : {:.1f}\nMAPE : {:.1f}%'.format(mse,mape))
```
